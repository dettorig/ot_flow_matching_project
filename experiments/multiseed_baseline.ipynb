# Multi-seed baseline

Run the four canonical couplings across multiple seeds and aggregate
mean ± std on key metrics.

```python
import matplotlib.pyplot as plt
import numpy as np
import yaml

from otfm.sweeps import SweepConfig, aggregate, run_baseline_sweep
```

```python
cfg = yaml.safe_load(open('../configs/base.yaml'))
sc = SweepConfig(
    n=cfg['multiseed']['n'],
    train_steps=cfg['multiseed']['train_steps'],
    widths=cfg['model']['widths'],
    seeds=cfg['multiseed']['seeds'],
)
df = run_baseline_sweep(sc)
df.head()
```

```python
metrics = ['NPE', 'endpoint_W2_sq_at_32', 'target_velocity_kernel_var', 'vel_diff_sq']
agg = aggregate(df, group_cols=['coupling'], metrics=metrics)
agg
```

```python
fig, axes = plt.subplots(2, 2, figsize=(11, 7))
for ax, m in zip(axes.ravel(), metrics, strict=True):
    d = agg.sort_values('coupling')
    ax.bar(d['coupling'], d[f'{m}_mean'], yerr=d[f'{m}_std'], capsize=4)
    ax.set_title(m)
    ax.tick_params(axis='x', rotation=30)
fig.tight_layout()
plt.show()
```
