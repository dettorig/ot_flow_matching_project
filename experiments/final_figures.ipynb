# Final figures (reads ``results/full_v1.csv``)

This notebook is the source of the figures shipped in the report.
It reads the CSV produced by ``scripts/run_full_benchmark.py`` and
produces clean publication-style plots.

```python
import pathlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from otfm.plotting import color_for, style_axes
```

```python
csv_path = pathlib.Path('../results/full_v1.csv')
if not csv_path.exists():
    raise FileNotFoundError(
        'Run scripts/run_full_benchmark.py first to populate results/full_v1.csv'
    )
df = pd.read_csv(csv_path)
df.head()
```

```python
# Headline figure: bar chart of NPE per coupling (mean ± std over seeds).
agg = df.groupby('coupling').agg(
    NPE_mean=('NPE', 'mean'),
    NPE_std=('NPE', 'std'),
    W2_mean=('endpoint_W2_sq_at_32', 'mean'),
    W2_std=('endpoint_W2_sq_at_32', 'std'),
).reset_index()
agg = agg.sort_values('coupling')

fig, axes = plt.subplots(1, 2, figsize=(11, 4))
for ax, (col_mean, col_std, title) in zip(
    axes,
    [('NPE_mean', 'NPE_std', 'NPE'), ('W2_mean', 'W2_std', r'endpoint $W_2^2$')],
    strict=True,
):
    colors = [color_for(c) for c in agg['coupling']]
    ax.bar(agg['coupling'], agg[col_mean], yerr=agg[col_std], capsize=4, color=colors)
    ax.tick_params(axis='x', rotation=30)
    style_axes(ax, title=title)
fig.tight_layout()
fig.savefig('../report/fig_headline.pdf', bbox_inches='tight')
plt.show()
```
